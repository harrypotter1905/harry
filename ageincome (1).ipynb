```python
incomes = df['Income'].dropna().values
n = len(incomes)
mean = sum(incomes) / n
sorted_incomes = sorted(incomes)
median = sorted_incomes[n // 2] if n % 2 else (sorted_incomes[n // 2 - 1] + sorted_incomes[n // 2]) / 2
min_val = min(incomes)
max_val = max(incomes)
std_dev = (sum((x - mean) ** 2 for x in incomes) / n) ** 0.5
```

```python
print(f"mean: {mean: .2f}")
print(f"median: {median: .2f}")
print(f"min: {min_val: .2f}")
print(f"max: {max_val: .2f}")
print(f"std: {std_dev: .2f}")
```

```python
print("\n By age group")
for age_group, group in df.groupby('Age'):
    values=group['Income'].dropna().values
    n = len(values)
    mean = sum(values)/n
    sorted_values = sorted(values)
    median = sorted_values[n//2] if n%2 else (sorted_values[n//2 -1] + sorted_values[n//2]) / 2
    min_valu = min(values)
    max_valu = max(values)
    std_dev = (sum((x - mean) ** 2 for x in values) / n) ** 0.5

    print(f"{age_group}")
    print(f"mean: {mean: .2f}")
    print(f"median: {median: .2f}")
    print(f"min: {min_valu: .2f}")
    print(f"max: {max_valu: .2f}")
    print(f"std: {std_dev: .2f}")
```
