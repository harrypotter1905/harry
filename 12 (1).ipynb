```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load from local CSV file
df = pd.read_csv("iris.csv")

# Display first few rows to verify
print(df.head())
```

```python
print("\nFeature Types:\n")
print(df.dtypes)

# Identify types manually
for col in df.columns:
    print(f"{col}: {'Nominal' if df[col].dtype == 'object' else 'Numeric'}")

```

```python
df.drop('Species', axis=1).hist(figsize=(10, 6), bins=20, color='lightblue', edgecolor='black')
plt.suptitle("Histograms of Features", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
for i, col in enumerate(df.columns[:-1], 1):  # Exclude species
    plt.subplot(1, 4, i)
    sns.boxplot(y=df[col], color='lightgreen')
    plt.title(col)
plt.tight_layout()
plt.suptitle("Boxplots of Features", fontsize=16)
plt.show()

```

```python
for feature in df.columns[:-1]:
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[feature] < lower) | (df[feature] > upper)]
    print(f"\nOutliers in {feature}:")
    print(outliers[[feature, 'Species']])
```
